```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# distribution values from Kupper flow-based cattle manure emissions and effect of anaerobic digestion

s = 1000
nh3_n_flow = np.random.triangular(14,16,19, size=s)
nh3_ad_eff = np.random.normal(1.59, 0.64, size=s)
nh3_n_ad = np.multiply(nh3_n_flow, nh3_ad_eff)
nh3_mean = np.mean(nh3_n_ad)
nh3_std = np.std(nh3_n_ad)

n2o_n_flow = np.random.triangular(0.08,0.13,0.18, size=s)
n2o_ad_eff = np.random.normal(1.16, 0.29, size=s)
n2o_n_ad = np.multiply(n2o_n_flow, n2o_ad_eff)
n2o_mean = np.mean(n2o_n_ad)
n2o_std = np.std(n2o_n_ad)

print('mean, std')
print(f'NH3: {nh3_mean:.2f}, {nh3_std:.2f}')
print(f'N2O: {n2o_mean:.2f}, {n2o_std:.2f}')


```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6.4,3.3))
ax1.hist(nh3_n_ad, bins=20, density=True)
ax2.hist(n2o_n_ad, bins=20, density=True)

ax1.set_title('NH3-N')
ax1.set_xlabel('%TAN')
ax2.set_title('N2O-N')
ax2.set_xlabel('%N')
fig.tight_layout()
```
